```python
import time
import json
def buyInventory():
        print('----buy item from inventory ----')
        print("*******************MENU*******************")
        print("product_id    product_name",end="\n")
        print("..........................................")
        for i in records.keys():
            print(i,"        ",records[i]["name"])  
        n=int(input("enter the no of items you want to purchase: "))
        s=0
        for i in range(n):
            prod_id=input("enter the product id : ")
            quant=int(input("enter the quantity: "))
            product = records.get(prod_id)
            if product:
                if(quant<=records[prod_id]['qn']):
                
                    print("Product-",records[prod_id]['name'])
                    print("Category-",records[prod_id]['Category'])
                    print("Price-",records[prod_id]['pr'])
                    print("Quantity-",quant)
                    print("Expiry Date-",records[prod_id]['Expiry Date'])
                    print("-------------------------------------------") 
                    print("time of purchase -", time.ctime())     
                    print("********************************************") 
                    p=records[prod_id]['pr']*quant
                    print("your total amount is = ",p)
                    print("********************************************") 
                    records[prod_id]["qn"]=records[prod_id]['qn']-quant
                    s=s+p
                    
            else:
                    print("invalid id")
        else:
                y=quant-records[prod_id]['qn']
                print("you have asked",y ,"more then we have in stock")
        
        print("you total purchase amount is ",s)

        CHOICE = int(input('Enter 98 to continue or 9 to exit: '))
        if CHOICE == 98:
            menuDisplay()
        else:
            print ("Thanks for visiting....")
            
```
